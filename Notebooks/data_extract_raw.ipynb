```python
#-- Lib import
import pandas as pd
import datetime as dt
import time
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from scipy import stats


#-- Lista de empresas
empresa = {
    'Apple': 'AAPL',
    'Microsoft': 'MSFT',
    'Google': 'GOOG',
    'Amazon': 'AMZN',
    'Tesla': 'TSLA',
    'NVIDIA': 'NVDA',
    'Meta': 'META'
}

#-- Download data
df_list = {}
for name,ativo in empresa.items():
  teste = yf.Ticker(ativo)
  teste = teste.history(interval = '1d',
                        start = '2000-01-01',
                        end = '2023-08-01')
  teste['Empresa'] = name
  teste['Symbol'] = ativo
  teste.reset_index(inplace=True)
  teste['Date'] = pd.to_datetime(pd.to_datetime(teste['Date']).dt.strftime('%Y-%m-%d'))
  teste.to_csv(f'../Dataset/Raw/{name.lower()}_data.csv',index=False)
  df_list[ativo] = teste


new_df = pd.concat(df_list.values())
new_df.reset_index(drop=True,inplace=True)
new_df.to_csv(f'../Dataset/Raw/full_data.csv',index=False)
new_df
```
